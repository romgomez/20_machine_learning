```python
#Import dependencies
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Import warnings to eliminate warning error when testing classifiers 
def warn(*args, **kwargs): pass
import warnings
warnings.warn = warn

# Importing classifiers
from sklearn.metrics import accuracy_score, log_loss
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier as RFC
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("cumulative.csv")
df = df.drop(columns=["rowid", "kepid", "kepoi_name", "kepler_name", "koi_pdisposition", "koi_score", "koi_tce_delivname"])
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
data = df.drop("koi_disposition", axis=1)
data.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = data
y = df["koi_disposition"].values.reshape(-1, 1)

print(X.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from keras.utils import to_categorical

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)


# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)

# Step 2: One-hot encoding
one_hot_y = to_categorical(encoded_y)
one_hot_y
```

```python
#Preliminary testing of classifiers fit to data (not scaled)
classifiers = [
    LogisticRegression(),
    KNeighborsClassifier(3),
    SVC(kernel="rbf", C=0.025, probability=True),
    DecisionTreeClassifier(),
    RandomForestClassifier()]

# Logging for Visual Comparison
log_cols=["Classifier", "Accuracy", "Log Loss"]
log = pd.DataFrame(columns=log_cols)

for clf in classifiers:
    clf.fit(X_train, y_train)
    name = clf.__class__.__name__
    
    print("="*30)
    print(name)
    
    print('****Results****')
    train_predictions = clf.predict(X_test)
    acc = accuracy_score(y_test, train_predictions)
    print("Accuracy: {:.4%}".format(acc))
    
    train_predictions = clf.predict_proba(X_test)
    ll = log_loss(y_test, train_predictions)
    
    log_entry = pd.DataFrame([[name, acc*100, ll]], columns=log_cols)
    log = log.append(log_entry)
    
print("="*30)
```

```python
# Plotting model accuracy for a few classifiers
sns.set_color_codes("muted")
sns.barplot(x='Accuracy', y='Classifier', data=log, color="b")

plt.xlabel('Accuracy %')
plt.title('Classifier Accuracy')
plt.show()
```

# Train the Support Vector Machine

```python
# Train support vector machine linear classifier, fit to the training data, and validate with the test data
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model
model.fit(X_train_scaled, y_train)
```

```python
# SVM Model Accuracy
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

# Logistic Regression


```python
# Train and test Logistic Regression model
from sklearn.linear_model import LogisticRegression
log_reg_model = LogisticRegression()
log_reg_model
log_reg_model.fit(X_train_scaled, y_train)
```

```python
# Logisitics Regression Model Accuracy
print(f"Training Data Score: {log_reg_model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {log_reg_model.score(X_test_scaled, y_test)}")
```

# Random Forest Classifier


```python
from sklearn.svm import SVC 
rfc_model = RFC()
rfc_model
rfc_model.fit(X_train_scaled, y_train)

# Logisitics Regression Model Accuracy
print(f"Training Data Score: {rfc_model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {rfc_model.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the `C` and `gamma` parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10],
              'gamma': [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
# Fit the model using the grid search estimator. 
# This will take the SVC model and try each combination of parameters
grid.fit(X_train_scaled, y_train)
```

```python
#Determine the best parameters and highest accuracy
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test)
predictions
```

```python
# Calculate classification report for SVC hypertuned model
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["CONFIRMED", "FALSE POSITIVE","CANDIDATE"]))
```
